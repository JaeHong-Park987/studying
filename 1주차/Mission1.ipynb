````python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/healthexp.csv")
df
````

````python
df1 = df[df['Year']>=2011] # 2011년이상의 값으로 된 새로운 데이터프레임 생성
df1 = df1.dropna() # 요약 통계량을 구하기 전 있을 수도 있는 결측값 제거
df1
````

````python
# groupby함수를 이용해서 구하기

df1.groupby(["Year","Country"])["Life_Expectancy"].mean().unstack()
````

````python
# pivot_table 이용해서 구하기

import pandas as pd
pd.pivot_table(df1,index = ["Year"],columns = ['Country'], values="Life_Expectancy", aggfunc = "mean")
````
