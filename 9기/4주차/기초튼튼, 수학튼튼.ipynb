## 딥러닝 학습방법 이해하기

```python
def softmax(vec):
    denumerator = np.exp(vec - np.max(vec, axis=-1, keepdims=True))
    numerator = np.sum(denumerator, axis = -1, keepdims = True)
    val = denumerator / numerator
    return val
```

```python
vec = np.array([[1,2,0],[-1,0,1],[-10,0,10]])
vec
```

```python
softmax(vec)
```

```python
def one_hot(val, dim):
    return [np.eye(dim)[_] for _ in val]

def one_hot_encoding(vec):
    vec_dim = vec.shape[1]
    vec_argmax = np.argmax(vec, axis = -1)
    return one_hot(vec_argmax, vec_dim)

def softmax(vec):
    denumerator = np.exp(vec - np.max(vec, axis=-1, keepdims=True))
    numerator = np.sum(denumerator, axis = -1, keepdims = True)
    val = denumerator / numerator
    return val
```

```python
vec = np.array([[1,2,0],[-1,0,1],[-10,0,10]])
print(one_hot_encoding(vec))
print(one_hot_encoding(softmax(vec)))
```

## 확률론 맛보기

* 딥러닝은 확률론 기반의 기계학습 이론에 바탕을 두고 있다.
* L2노름은 예측 오차의 분산을 가장 최소화 하는 방향으로 학습하도록 유도
* 교차 엔트로피는 모델 예측의 불확실성을 최소화하는 방향으로 학습하도록 유도
* 분산, 불확실성을 최소화하기 위해 측정하는 방법을 알아야 한다.

* 확률변수는 확률분포에 따라 이산형, 연속형으로 구분된다.
    * 이산형 : 확률변수가 가질 수 있는 경우의 수를 모두 고려하여 확률을 더해서 모델링한다.
    * 연속형 : 데이터 공간에 정의된 확률변수의 밀도 위에서의 적분을 통해 모델링한다.

* 조건부 확률 
    * 이산형 : 확률을 의미
    * 연속형 : 밀도를 의미
* 로지스틱 회귀에서 사용했던 선형모델과 소프트맥스 함수의 결합은 데이터에서 추출된 패턴을 기반으로 확률을 해석하는데 사용

* 기대값
    * 통계적 범함수를 계산할 수 있다.
    * 기대값은 데이터를 대표하는 통계량이다.
    * 기대값을 이용해 분산, 첨도, 공분산 등 여러 통계량을 계산할 수 있다.

* 몬테카를로 샘플링
    * 확률분포를 모를 때 데이터를 이용하여 기대값을 계산하려면 몬테카를로 샘플링 방법을 사용한다.
    * 독립추출만 보장된다면 대수의 법칙에 의해 수렴성을 보장한다.

image.png

```python
import numpy as np

def mc_int(fun, low, high, sample_size = 100, repeat = 10):
    int_len = np.abs(high - low)
    stat = []
    for _ in range(repeat):
        x = np.random.uniform(low = low, high = high, size = sample_size)
        fun_x = fun(x)
        int_val = int_len * np.mean(fun_x)
        stat.append(int_val)
    return np.mean(stat), np.std(stat) # 평균, 분산(오차 범위)

def f_x(x):
    return np.exp(-x**2)

print(mc_int(f_x, low = -1, high = 1, sample_size = 10000, repeat = 100))
```

## 통계학 맛보기

* 통계적 모델링은 적절한 가정 위에서 확률분포를 추정하는 것
* 근사적으로 확률분포를 추정할 수 밖에 없다.
    * 정확하게 맞추는 것보다는 데이터와 추정 방법의 불확실성을 고려해서 위험을 최소화하는 것이다
* 데이터가 특정 확률 분포를 따른다고 가정한 후 그 분포를 결정하는 모수를 추정하는 방법을 모수적 방법론이라 한다.
* 확률분포를 가정하지 않고 모델의 구조 및 모수의 개수가 유연하게 바뀌면 비모수 방법론이라 한다.
    * 비모수 : 모수를 안 쓰는 것이 아님!!

* 확률분포 가정
    * 데이터가 2개의 값(0,1) -> 베르누이
    * 데이터가 n개의 이산적인 값 -> 카테고리
    * 데이터가 0~1사이 값 -> 베타
    * 데이터가 0이상의 값 -> 감마, 로그정규
    * 데이터가 실수 전체 값 -> 정규, 라플라스
* 데이터를 생성하는 원리를 먼저 고려하는 것이 원칙!!!
    * 기계적으로 확률분포를 가정해서는 안된다.

* 통계량(표본평균, 표본 분산)의 확률분포를 표집분포라고 부른다. (표본분포와는 다름)
    * 표집분포는 N이 커질수록 정규분포를 따른다 : 중심극한정리

* 가장 가능성이 높은 모수를 추정하는 방법 중 하나는 최대가능도 추정법이다.
    * 데이터가 독립적으로 추출될 경우 로그가능도를 최적화합니다.
        * 데이터의 수가 수억 단위가 된다면 컴퓨터의 정확도로는 가능도를 계산하는 것은 불가능하다.
        * 로그를 사용하면 덧셈으로 바꿀 수 있기 때문에 연산이 가능해진다.
        * 연산량도 줄어든다.

* 확률분포의 거리를 구해보자
    * 기계학습에서 사용되는 손실함수들은 모델이 학습하는 확률분포와 데이터에서 관찰되는 확률분포의 거리를 통해 유도합니다.
    * 두 확률분포 사이의 거리를 계산할 때 사용하는 함수
        * 총변동 거리(TotalVariationDistance,TV)
        * 쿨백-라이블러발산(Kullback-LeiblerDivergence,KL)
        * 바슈타인거리(WassersteinDistance)

* 확률과 가능도의 차이는 무엇일까?
    * 개념적 차이
        * 확률 : 주어진 확률분포에서 해당 관측값이 나올 확률
        * 가능도 : 주어진 관측값이 특정 확률분포로부터 나왔을 확률(= 연속확률밀도함수의 y값)
    * 수식적 차이
        * 확률 : P(관측값 X | 확률분포 D)
        * 가능도 : L(확률분포 D | 관측값 X)
    * 확률밀도함수에서의 차이

* 확률 대신 가능도를 사용하였을 때의 이점은 어떤 것이 있을까?
    * 이론적으로 가장 가능성이 높은 모수를 추정할 수 있다.



```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.arange(0, 1, 0.001)
p = theta ** 3 * (1 - theta) ** 7
plt.plot(theta, p)
plt.show()
```

* n = 10, k=3, p = theta인 이항분포를 나타냅니다.
* K ~ B(10,theat)
