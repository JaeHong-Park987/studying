```python
import pandas as pd

idx = ["HDD", "SSD", "USB", "CLOUD"]
data = [19, 11, 5, 97]

#위 데이터로 Series를 구현해보세요.
series = pd.Series(data = data , index = idx)

# 10 이상 20 이하를 가지는 데이터만 이용해 다시 series를 정의하세요.
series = series[(series >= 10) & (series <= 20)]

print(series)
```

# 2

```python
import pandas as pd

df1 = pd.DataFrame({'Name': ['cherry', 'mango', 'potato', 'onion'],
                    'Type': ['Fruit', 'Fruit', 'Vegetable', 'Vegetable'],
                    'Price': [100, 110, 60, 80]})
df2 = pd.DataFrame({'Name': ['pepper', 'carrot', 'banana', 'kiwi'],
                    'Type': ['Vegetable', 'Vegetable','Fruit', 'Fruit'],
                    'Price': [50, 70, 90, 120]})

#df1, df2를 colums를 이용해 결합
df3= pd.concat([df1, df2])
df3

# Fruit와 Vegetable의 type에 따라 정렬하고, 가격을 내림차순으로 정리 
df_fruit = df3[df3['Type'] == 'Fruit'].sort_values(by = 'Price', ascending = False)
df_veg = df3[df3['Type'] == 'Vegetable'].sort_values(by = 'Price', ascending = False)

# Fruit와 Vegetable 상위 2개의 가격의 합을 출력 
print("Sum of Top 2 Fruit Price :", sum(df_fruit[:2]['Price'])) 
print("Sum of Top 2 Vegetable Price :", sum(df_veg[:2]['Price']))
```

```python
import pandas as pd

idx = ["Sue", "Ryan", "Jay", "Jane", "Anna"]
col = ["round_1", "round_2", "round_3", "round_4", "round_5"] 
data= [[55, 65, 60, 66, 57],
            [64, 77, 71, 79, 67],
            [88, 81, 79, 89, 77],
            [45, 35, 30, 46, 47],
            [91, 96, 90, 97, 99]]

#위 데이터를 이용해 dataframe을 구성해보세요.
df = pd.DataFrame(data, index = idx, columns = col)

#df에 새로운 column인 round_6의 데이터 [11, 15, 13, 17, 19]를 추가해보세요. 
df['round_6'] = [11, 15, 13, 17, 19]
print(df)

#각 데이터의 mean, max, min 값을 구해 출력해보세요. 
print(df.describe().loc[['mean', 'max', 'min']])
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 5., 0.2)

# 사용 마커는 다음과 같습니다. red dashes, green suares, blue triangles
plt.figure(figsize = (8,5))
plt.plot(t, t, color = 'r', linestyle = 'dashed')
plt.plot(t, t ** 2, 'gs')
plt.plot(t, t ** 3, 'b^') 
plt.show()
```
