```python
def swap_value(x,y):
    temp = x
    x = y
    y = temp

def swap_reference(list_ex, offset_x, offset_y):
    temp = list_ex[offset_x]
    list_ex[offset_x] = list_ex[offset_y]
    list_ex[offset_y] = temp 

ex = [1,2,3,4,5]
swap_value(ex[0], ex[1])
print(ex) # 바뀌는 것 x

swap_reference(ex, 3, 4)
print(ex) # 바뀜
```

```python
# generator -> 용량을 아낄때 사용!!

def general_list(value):
    result = []
    for i in range(value):
        result.append(i)
    return result

print(general_list(50))
```

```python
import sys

result = general_list(50)
sys.getsizeof(result)
```

```python
def generator_list(value):
    result = []
    for i in range(value):
        yield i

print(generator_list(50)) # 메모리만 던져준다.

for a in generator_list(50):
    print(a)

print(sys.getsizeof(generator_list(50))) # 112메모리만 사용하여 더 절약 가능    
```

```python
# keyword arguments

def print_somthing(my_name, your_name, third_name):
    print("Hello {0}, My name is {1}".format(your_name, my_name))

print_somthing(third_name = 'abc', my_name='JH', your_name = 'SW')

# default arguments
def print_somthing(my_name, your_name = 'SW'):
    print("Hello {0}, My name is {1}".format(your_name, my_name))

print_somthing('JH')
print_somthing('JH', 'NAVER')
```

```python
# Asterisk -> 개수가 정해지지 않은 변수

def asterisk_test(a, b, *args):
    return a + b + sum(args)

print(asterisk_test(1,2,3,4,5))
```

```python
# 키워드 가변인자 -> **를 사용(dict type)

def kwargs_test_1(**kwargs):
    print(kwargs)

kwargs_test_1(first = 3, second = 4)
```

```python
def kwargs_test(one, two, *args, ** kwargs):
    print(one+two+sum(args))
    print(kwargs)
kwargs_test(10, 30, 3,5,6,7,8,9, first = 3, second = 4)
```

```python
# unpacking a container

def asterisk_test(a, *args): # * -> 여러개의 값을 가질 수 있게 해주는 의미
    print(a, args)
    print(type(args))

asterisk_test(1, *(2,3,4,5,6)) # * -> unpacking
```
