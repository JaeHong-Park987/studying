```python
num_list = [1, 5, 7, 15, 16, 22, 28, 29]

def get_odd_num(num_list):
    new_list = []
    for i in num_list:
        # if i % 2 != 0:
        if i % 2 == 1:
            new_list.append(i)
    
    return new_list


print(get_odd_num(num_list))   
```

```python
# 한줄로 표현하기 위해 lambda 함수 활용

def get_odd_num(num_list):   
    return list(filter(lambda x: x % 2 == 1, num_list))


print(get_odd_num(num_list))   
```

```python
sentence = 'way a is there will a is there Where'

def reverse_sentence(sentence):
    new_sentence = sentence.split(' ')[::-1] # 역순으로 한 단어씩 받아오기
    return ' '.join(new_sentence) # join 함수 : 리스트를 string으로 변환해주는 함수 -> 연결자.join(연결할 리스트)

print(reverse_sentence(sentence))
```

```python
def reverse_sentence(sentence):
    new_sentence = sentence.split(' ')[::-1]
    new = "" # 결과를 담을 스트링 변수 생성
    for i in new_sentence:
        new += i + " " 
    return new.rstrip() # 마지막 공백 제거를 위해 rstrip()활용

print(reverse_sentence(sentence))
```

```python
score = [(100, 100), (95, 90), (55, 60), (75, 80), (70, 70)]

def get_avg(score):
    for i in range(len(score)):
        print('{} 번, 평균 : {}'.format(i + 1, (score[i][0] + score[i][1]) / 2))

get_avg(score)
```

```python
dict_first = {'사과' : 30, '배' : 15, '감' : 10, '포도' : 10}
dict_second = {'사과' : 5, '감' : 25, '배' : 15, '귤' : 25}

def merge_dict(dict_first, dict_second):
    for i in dict_second:
        if i in dict_first: # dict_second의 키 값이 dict_first 에도 있으면 
            dict_first[i] += dict_second[i] # 두 value를 합해준다.
        else: # 없으면
            dict_first[i] = dict_second[i] # dict_first에 dict_second 값을 추가해준다.
    return dict(sorted(dict_first.items())) # 오름차순으로 정렬


print(merge_dict(dict_first, dict_second))
```

```python
inputs = 'cat32dog16cow5'

def find_string(inputs):
    inputs = list(inputs)
    for i in range(len(inputs)):
        if inputs[i].isdigit():
            inputs[i] = ' '
    new_word = ""
    new_list = []        
    for i in inputs:
        if i != ' ':
            new_word += i
        elif new_word != '':
            new_list.append(new_word)
            new_word = ""
    
    return new_list
        


string_list = find_string(inputs)
print(string_list)
```

```python
inputs = 'cat32dog16cow5'

def find_string(inputs):
    new_word = ''
    new_list = []

    for i in inputs :
        if not i.isdigit() :
            new_word += i
        elif new_word != '' :
            new_list.append(new_word)
            new_word = ''
    return new_list


print(find_string(inputs))
```
