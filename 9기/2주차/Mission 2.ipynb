```python
class Score():
    def __init__(self, mid, final):
        self.__mid = mid # 중간고사 인스턴스 private으로 선언
        self.__final = final # 기말고사 인스턴스 private으로 선언

    @property
    def mid(self):
        return self.__mid

    @property
    def final(self):
        return self.__final
    

score = Score(50, 75)
print((score.mid + score.final) / 2)
```

```python
class Car():

    def __init__(self,fuel,wheels):
        self.fuel = fuel
        self.wheels = wheels

class Bike(Car): # 부모 클래스 Car으로 부터 상속
    def __init__(self, fuel,wheels,size):
        super().__init__(fuel,wheels) # 부모객체 사용
        self.size = size # 속성 값 추가


bike = Bike("gas", 2, "small")
print(bike.fuel, bike.wheels, bike.size)
```

```python
file_path = '/content/data-01-test-score.csv'

class ReadCSV():
    def __init__(self, file_path):
        self.file_path = file_path
    
    def read_file(self):
        line_counter = 0 # 파일의 총 줄 수를 세는 변수
        score_list = [] # 학생별 score를 저장하는 list

        with open(file_path) as my_file:
            while True:
                data = my_file.readline().replace("\n","") # 한줄씩 불러오면서 줄바꿈 제거
                if not data : break
                else:
                    score_list.append(data.split(","))
                line_counter += 1

        for i in range(len(score_list)):
            score_list[i] = list(map(int, score_list[i]))
            i += 1
        
        self.score_list = score_list

        return score_list
    
    def merge_list(self):
        self.read_file()
        total_list = []
        for i in range(len(self.score_list)):
            total_list.append(sum(self.score_list[i]) / len(self.score_list[i]))
        total_list.sort()
        return total_list




read_csv = ReadCSV(file_path)
print(read_csv.merge_list())
```
