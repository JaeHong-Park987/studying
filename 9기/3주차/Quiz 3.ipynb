# Q1

```python
import numpy as np

x = np.array([-1,2,-3])

def l1_norm(x):
    x_norm = np.abs(x) # 절대값!!
    x_norm = np.sum(x_norm) # 원소 더하기
    return x_norm

l1_norm(x)
```

* L1 : 각 성분의 변화량의 절대값을 모두 더합니다. -> 1 + 2 + 3 = 6

# Q2

```python
x = np.array([-6,-8])

def l2_norm(x):
    x_norm = x*x # 제곱
    x_norm = np.sum(x_norm) # 제곱 합
    x_norm = np.sqrt(x_norm) # 루트처리
    return x_norm

l2_norm(x)
```

* L2 : 피타고라스 정리를 이용해 유클리드 거리를 계산 -> 36 + 64 = 100에 루트 -> 10

# Q3

```python
x = np.array([-3,0])
y = np.array([0,4])

distance = np.abs(y - x)

l2_norm(distance)
```

# Q4

```python
x = np.array([0,1])
y = np.array([0,2])

def angle(x, y):
    v = np.inner(x,y) / (l2_norm(x) * l2_norm(y))
    theta = np.arccos(v)
    return theta

angle(x, y)
```
