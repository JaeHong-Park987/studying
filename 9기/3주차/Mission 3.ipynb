```python
import numpy as np

arr1 = np.random.rand(5,3) # 무작위의 데이터를 가진 5*3 행렬 생성 
arr2 = np.random.rand(3,2) # 무작위의 데이터를 가진 3*2 행렬 생성

dot = arr1 @ arr2 # 행렬곱 연산 -> 예시코드와는 다른 방법

print(dot, dot.shape)
```

```python
arr1 = np.array([[5,7], [9,11]], float)
arr2 = np.array([[2,4],[6,8]], float)

# concatenate 연산
concat_1 = np.concatenate((arr1, arr2), axis = 0) # 행방향으로 더하기
concat_2 = np.concatenate((arr1, arr2), axis = 1) # 열방향으로 더하기

print(concat_1)
print(concat_2)
```

* 행방향으로 더하기 -> 데이터가 밑으로 더해진다.
* 열방향으로 더하기 -> 데이터가 옆(오른쪽)으로 더해진다.

```python
import numpy as np

xy = np.array([[1.,2.,3.,4.,5.,6.],
               [3.,6.,9.,12.,15.,18.]])

# 경사하강법을 위한 데이터 분리
x_train = xy[0] # xy의 첫번째 행 데이터 가져오기
y_train = xy[1] # xy의 두번째 행 데이터 가져오기

# array형태로 출력하기 위해 파라미터에 1 추가
beta_gd = np.random.rand(1) # weight : 직선의 기울기(임의의 값)
bias = np.random.rand(1) # bias : y 절편(임의의 값)


learning_rate = 0.01

for i in range(1000):
    error = np.sum((y_train - (x_train * beta_gd + bias))**2) / len(y_train) # 차이의 제곱을 이용해서 계산 -> 루트 필요없음(sqrt x)
    grad_beta_gd = - (2 / len(y_train)) * np.sum((y_train - x_train*beta_gd - bias) @ x_train.T)
    grad_bias = - (2 / len(y_train)) * np.sum(y_train - (x_train * beta_gd + bias))

    beta_gd = beta_gd - 0.01 * grad_beta_gd
    bias = bias - 0.01 * grad_bias


    if i % 100 == 0:
        print('Epoch ({:10d}/1000) error: {:10f}, beta_gd : {:10f}, bias:{:10f}'.format(i, error, beta_gd.item(), bias.item()))
```
