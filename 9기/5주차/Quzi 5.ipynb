# Q10

* softmax함수
    * 함수에 각각 입력된 값은 0~1사이 값으로 출력된다.
    * softmax 출력의 총합은 1이다.
    * softmax 함수를 적용해도 입력 값의 대소 관계는 변하지 않는다.
    * softmax를 이용해 이진분류 모델을 구성할 수 있다.
        * 보통 다중분류에서 사용되지만 이진 분류도 가능하다.
