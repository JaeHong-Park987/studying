* 동물병원은 `경기도, 서울특별시, 인천광역시, 경상북도` 순으로 많았다.
* 약국은 `경기도, 서울특별시, 부산광역시, 경상남도` 순으로 많았다.
* 병원은 `경기도,서울특별시, 부산광역시, 경상남도` 순으로 많았다.

* 결론 : 경기도, 서울특별시에는 병원이 많고 약국, 동물병원 또한 많다. 그러나 동물병원 같은 경우에는 약국이 많은 부산광역시, 경상남도와 다르게 인천광역시와 경상북도에 많다.


#3.

```python
# 서울에 소재한 병원중 상권업종소분류명이 피부, 성형이 들어간 데이터

df_face = df[(df["시도명"] == "서울특별시") & (df["상권업종중분류명"] == "병원") & (df["상권업종소분류명"].str.contains("피부|성형"))]
df_face
```

```python
# 서울특별시 시군구명별 피부,성형이 들어간 데이터 수
face_n = df_face["시군구명"].value_counts()
face_n
```

```python
# 서울에 소재한 병원 데이터
df_seoul = df[(df["시도명"] == "서울특별시") & (df["상권업종중분류명"] == "병원")]
df_seoul
```

```python
# 서울 시군구명별 병원 수
seoul_n = df_seoul["시군구명"].value_counts()
seoul_n
```

```python
# face_n : 피부과 성형외과 수 / seoul_n : 병원 수
# round 함수를 이용해 소수점 두 번째짜리까지 출력


round(face_n/seoul_n,2).sort_values(ascending = False)
```

```python
# 노인/치매병원 데이터

df_old = df[df["상권업종소분류명"] == "노인/치매병원"]
df_old
```

```python
# 시도명에 결측값이 있지만 경도 위도 값은 모두 있기 때문에 시각화하는데 모든 데이터를 사용해도 될 것 같다.

df_old.info()
```

```python
import folium

print(df_old["위도"].mean())
print(df_old["경도"].mean())
```

```python
map = folium.Map(location = [df_old["위도"].mean(), df_old["경도"].mean()],zoom_start = 7)


for n in df_old.index:
  name = df_old.loc[n,"상호명"]
  address = df_old.loc[n,"도로명주소"]
  popup = f"{name}-{address}"
  location = [df_old.loc[n,"위도"],df_old.loc[n,"경도"]]
  folium.Marker(
      location = location, 
      popup = popup,
  ).add_to(map)

map
```
