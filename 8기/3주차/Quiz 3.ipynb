# 7

````python
df_sample = df.sample(1000,random_state = 1)

plt.figure(figsize= (15,4))
sns.lineplot(data = df_sample, x= "연령대코드(5세단위)", y = "신장(5Cm단위)", hue = "음주여부", ci = "sd")
````
